```python
train_dir = '/content/gdrive/MyDrive/Kaggle/fruits-360/Test/data_model/train'
val_dir = '/content/gdrive/MyDrive/Kaggle/fruits-360/Test/data_model/val'
```

```python
print('Total kelas Pelatihan:', len(os.listdir(train_dir)))
print('Total kelas Validasi:', len(os.listdir(val_dir)))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
                    rescale=1./255,
                    horizontal_flip=True,
                    shear_range = 0.4,
                    zoom_range=0.4,)

val_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
        train_dir, 
        target_size=(100, 100), 
        batch_size=32,
        class_mode='categorical')
validation_generator = val_datagen.flow_from_directory(
        val_dir, 
        target_size=(100, 100),
        batch_size=32, 
        class_mode='categorical')
```

```python
import tensorflow as tf

class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.95):
      print("\nAkurasi telah mencapai >90%!")
      self.model.stop_training = True
callbacks = myCallback()
```

```python
from keras.models import Sequential 
from keras.layers import Conv2D, MaxPooling2D, Activation, Dropout, Flatten, Dense


model = Sequential()
model.add(Conv2D(128, 3, activation="relu", input_shape=(100,100,3)))
model.add(MaxPooling2D())
model.add(Conv2D(64, 3, activation="relu"))
model.add(Conv2D(32, 3, activation="relu"))
model.add(MaxPooling2D())
model.add(Dropout(0.50))
model.add(Flatten())
model.add(Dense(5000, activation = "relu"))
model.add(Dense(1000, activation = "relu"))
model.add(Dense(131, activation = "softmax"))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer= "SGD",
              metrics=['accuracy'])
```

```python
history = model.fit(
      train_generator,
      steps_per_epoch=80, 
      epochs=100,
      validation_data=validation_generator, 
      validation_steps=80,
      callbacks=[callbacks])
```

akurasi telah mencapai >95%

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Akurasi Model')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss Model')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
```

```python
with tf.io.gfile.GFile('model.tflite', 'wb') as f:
  f.write(tflite_model)
```
